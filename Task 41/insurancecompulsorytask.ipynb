```python
# COMPULSORY TASK1 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline

import sklearn
from sklearn import linear_model
```

```python
insurance_data = pd.read_csv("insurance.csv")
insurance_data.head()
```

The target variable here is charges and remaining six variables such as age, sex, bmi, children, smoker, region are independent variable.

```python
x_age = insurance_data.loc[:,'age'].values.reshape(-1,1)
y_charges = insurance_data.loc[:,'charges'].values


#showing scatter plot with age on the x-axis and charges on the y-axis.
plt.scatter(x_age,y_charges,color = 'b')
plt.xlabel('Independent Variable -> Age')
plt.ylabel('Dependent Variable -> Charges')
plt.show()
```

```python
#using LinearRegression from sklearn, fit a model to your data, and make predictions on data.

cost_model = LinearRegression()
cost_model.fit(x_age,y_charges)
```

We can see how similar those predictions are to the x values we know the corresponding y values for. Below these predictions are shown in red, and the obversed values are in blue.

```python
y_pred = cost_model.predict(x_age)

plt.scatter(x_age,y_charges,color = 'b')
plt.plot(x_age,y_pred,color = 'r')
plt.ylabel('Charges')
plt.xlabel('Age')
plt.show()
```

It means that from the three independent variables, age is one of the factor, it affects the dependent variable (charges). It seems like while the rest are also factor may be the other variables outside the model because best line fit is not as expected. and doesnot seem to be growing in much progress.

```python
# testing more prediction making upto 70 age
unk_x = [[70]]

x_pred = np.append(x_age, unk_x).reshape(-1,1)
y_pred = cost_model.predict(x_pred)

plt.scatter(x_age,y_charges,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Insurance Costs')
plt.show()

print("Insurance costs:", cost_model.predict(unk_x))
```

# How BMI affects insurance costs

```python
x_bmi = insurance_data.loc[:,'bmi'].values.reshape(-1,1)
y_charges = insurance_data.loc[:,'charges'].values


#showing scatter plot with age on the x-axis and charges on the y-axis.
plt.scatter(x_bmi,y_charges,color = 'b')
plt.xlabel('Independent Variable -> BMI')
plt.ylabel('Dependent Variable -> Charges')
plt.show()
```

```python
#using LinearRegression from sklearn, fit a model to your data, and make predictions on data.

cost_model = LinearRegression()
cost_model.fit(x_bmi,y_charges)
cost_model.score(x_bmi,y_charges)
```

```python
y_pred = cost_model.predict(x_bmi)

plt.scatter(x_bmi,y_charges,color = 'b')
plt.plot(x_bmi,y_pred,color = 'r')
plt.ylabel('Charges')
plt.xlabel('BMI')
plt.show()
```

It means that from the three independent variables, BMI is one of the factor, it affects the dependent variable (charges) but not as expected. It seems like while the rest are also factor, may be the other variables outside the model because best line fit is not as expected and doesnot seem to be growing in much progress.
